```python
df = pd.read_csv('finish-crime-weather.csv')
```

```python
df['Universities_colleges_distance_mean'] = 0
df['Universities_colleges_distance_min'] = 0
```

```python
def distance(x1,y1,x2,y2):    
    return ((x1-x2)**2+(y1-y2)**2)**(1/2)
```

```python
i = 0

while i < len(df):
    
    k = 0
    
    array = []
    
    while k < len(df1):
        
        
        array.append(
            distance(
                df1.Latitude[k],
                df1.Longitude[k],
                df.Lat[i],
                df.Long[i]
            )
        )
        
        
        k += 1
    
    a = np.array(array)
    
    df['Universities_colleges_distance_mean'][i] = a.min()*100000
    df['Universities_colleges_distance_min'][i] = np.percentile(a, 25)*100000
    
    print(i)
    
    i += 1
```
