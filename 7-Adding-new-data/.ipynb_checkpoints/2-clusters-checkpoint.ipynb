```python
import numpy as np
import pandas as pd
```

```python
# visualization

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('data/finish-crime-weather.csv')
```

```python
df = df.drop('Unnamed: 0',1)
df = df.drop('Unnamed: 0.1',1)
```

```python
df.head()
```

```python
df['clust_50'] = 0
df['clust_100'] = 0
df['clust_200'] = 0
```

```python
model_50 = KMeans(50)
model_50.fit(df[['Lat','Long']].loc[(df.Lat > 40) & (df.Long < -70)])

df['clust_50'].loc[(df.Lat > 40) & (df.Long < -70)] = model_50.predict(
    df[['Lat','Long']].loc[(df.Lat > 40) & (df.Long < -70)]
)
```

```python
model_100 = KMeans(50)
model_100.fit(df[['Lat','Long']].loc[(df.Lat > 40) & (df.Long < -70)])

df['clust_100'].loc[(df.Lat > 40) & (df.Long < -70)] = model_100.predict(
    df[['Lat','Long']].loc[(df.Lat > 40) & (df.Long < -70)]
)
```

```python
model_200 = KMeans(50)
model_200.fit(df[['Lat','Long']].loc[(df.Lat > 40) & (df.Long < -70)])

df['clust_200'].loc[(df.Lat > 40) & (df.Long < -70)] = model_200.predict(
    df[['Lat','Long']].loc[(df.Lat > 40) & (df.Long < -70)]
)
```
