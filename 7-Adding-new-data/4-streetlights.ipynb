```python
df = df[['Lat', 'Long']]
```

```python
x = df['Long']
y = df['Lat']
```

```python
# Custom the inside plot: options are: scatter | reg | resid | kde | hex

sns.jointplot(x, y, kind='scatter')
sns.jointplot(x, y, kind='hex')
sns.jointplot(x, y, kind='kde')
```
