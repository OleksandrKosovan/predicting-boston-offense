```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv('data/Non_Public_Schools.csv')
```

```python
df1.head()
```

```python
df1 = df1[['X','Y']]
```

```python
df1.shape
```

```python
df1.describe()
```

```python
df = pd.read_csv('data/df-pschools-2.csv')
```

```python
df.head()
```

```python
df['Non-Public_schools_distance_25'] = 0
df['Non-Public_schools_distance_min'] = 0
df['Non-Public_schools_number_near'] = 0
```

```python
def distance(x1,y1,x2,y2):    
    return ((x1-x2)**2+(y1-y2)**2)**(1/2)
```

```python
i = 0

while i < len(df):
    
    k = 0
    
    array = []
    
    while k < len(df1):
        
        
        array.append(
            distance(
                df1.Y[k],
                df1.X[k],
                df.Lat[i],
                df.Long[i]
            )
        )
        
        
        k += 1
    
    n = 0
    number_near = 0
    
    
    while n < len(array):
        
        if (array[n]*100000) < 3000:
            number_near += 1
            n +=1
        else:
            n += 1
    
    
    a = np.array(array)
    
    df['Non-Public_schools_distance_min'][i] = a.min()*100000
    df['Non-Public_schools_distance_25'][i] = np.percentile(a, 25)*100000
    df['Non-Public_schools_number_near'][i] = number_near
    
    i += 1
```
