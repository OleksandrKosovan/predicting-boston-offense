# 3. Functions

```python
def create_list_number_crime(name_column, list_unique):
    # list_unique = df[name_column].unique()
    
    i = 0
    
    list_number = list()
    
    while i < len(list_unique):
        list_number.append(len(df.loc[df[name_column] == list_unique[i]]))
        i += 1
    
    return list_unique, list_number
```

```python
def bar_chart(list_number, list_unique):
    objects = list_unique
    y_pos = np.arange(len(objects))
    performance = list_number
 
    plt.figure(figsize=(20,10))    
    plt.bar(y_pos, performance, align='center', alpha=0.5)
    plt.xticks(y_pos, objects)
    plt.ylabel('Number') 
    plt.show()
    
    return 0
```

# 5. Create new features

I created 4 new weakly relevant features.

### 5.1. Day / Night

https://www.timeanddate.com/sun/usa/boston

**Values: 0/1**

- 1    418241 - DAY
- 0    156985 - NIGHT

**How did i create this feature?**

I used information about Yearly Sun Graph for Boston. Link:

https://www.timeanddate.com/sun/usa/boston 

And i created array which i used for select of day or night crimes. 

day_night_period = [
    [1,6,18],  [2,6,19], [3,6,20], [4,5,20],
    [5,5,21], [6,4,21], [7,5,21], [8,5,21],
    [9,6,20], [10,6,19], [11,6,17],[12,7,17]
]

0 - month / 1 - day start / 2 - night start

**Night**

*Values: 0/1*

- 0    418 241 - Day
- 1    156 985 - Night

**How did i create this feature?**


I did next simple steps:
(Day == 0) => Night = 1
(Day ==1) => Night = 0


```python
day_night_period = [
    [1,6,18],
    [2,6,19],
    [3,6,20],
    [4,5,20],
    [5,5,21],
    [6,4,21],
    [7,5,21],
    [8,5,21],
    [9,6,20],
    [10,6,19],
    [11,6,17],
    [12,7,17]
]
```

```python
df['Day'] = 0

i = 0

while i < 12:
    
    df['Day'].loc[
        (df['MONTH'] == day_night_period[i][0]) & 
        (df['HOUR'] >= day_night_period[i][1]) & 
        (df['HOUR'] <= day_night_period[i][2])
    ] = 1
    
    i +=1 
    
```

```python
df['Night'] = 0
df['Night'].loc[df['Day']==0]=1
```

```python
df['Day'].value_counts()
```

```python
df['Day'].value_counts().plot.bar()
plt.show()
```

```python
df['Night'].value_counts()
```

```python
df['Night'].value_counts().plot.bar()
plt.show()
```

## 5.2. ToNight

It is feature, which shows number of hours to night.

I used information about Yearly Sun Graph for Boston and i selected different values. Then i calculated it. 


```python
day_night_period
```

```python
df['ToNight'] = 0

i = 0

while i < 12:
    
    df['ToNight'].loc[
        (df['MONTH'] == day_night_period[i][0]) & 
        (df['Day']==1)
    ] = day_night_period[i][2] - df['HOUR'].loc[(df['MONTH'] == day_night_period[i][0]) & (df['Day']==1)] + 1
    
    i +=1 
 
```

```python
df['ToNight'].value_counts().plot.bar()
plt.show()
```

```python
df['ToNight'].value_counts()
```

## 5.3. ToDay

It is feature, which shows number of hours to day.

I used information about Yearly Sun Graph for Boston and i selected different values. Then i calculated it. 

```python
df['ToDay'] = 0
```

```python
i = 0 
while i < 12:
    
    
    df['ToDay'].loc[
        (df['HOUR'] < 8) &
        (df['MONTH'] == day_night_period[i][0]) &
        (df['Night']== 1)
    ] = day_night_period[i][1] - df['HOUR'].loc[(df['HOUR'] < 8) & (df['MONTH'] == day_night_period[i][0]) & (df['Night']==1)] + 1
    
    df['ToDay'].loc[
        (df['HOUR'] > 16) &
        (df['MONTH'] == day_night_period[i][0]) &
        (df['Night']== 1)
    ] = (24 - df['HOUR'].loc[(df['HOUR'] > 16) & (df['MONTH'] == day_night_period[i][0]) & (df['Night']== 1)]) + day_night_period[i][1] + 1

    
    i +=1
```

```python
df['ToDay'].value_counts()
```

```python
df['ToDay'].value_counts().plot.bar()
plt.show()
```
