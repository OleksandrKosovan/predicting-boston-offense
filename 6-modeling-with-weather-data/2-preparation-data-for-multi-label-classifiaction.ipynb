# Create new data set

```python
df_labels = pd.DataFrame()
```

```python
df_labels['DAY_OF_WEEK'] = 0
df_labels['DISTRICT'] = 0
df_labels['HOUR'] = 0
df_labels['MONTH'] = 0
df_labels['Day'] = 0
df_labels['Night'] = 0
df_labels['ToNight'] = 0
df_labels['ToDay'] = 0
df_labels['temperatureMin'] = 0
df_labels['temperatureMax'] = 0
df_labels['temperatureDifference'] = 0
df_labels['precipitation'] = 0
df_labels['snow'] = 0
```

```python
df_labels['Motor_Vehicle_Accident_Response'] = 0
df_labels['Larceny'] = 0
df_labels['Medical_Assistance'] = 0
df_labels['Simple_Assault'] = 0
df_labels['Violations'] = 0
df_labels['Investigate_Person'] = 0
df_labels['Vandalism'] = 0
df_labels['Drug_Violation'] = 0
df_labels['Larceny_From_Motor_Vehicle'] = 0
```

### Fucnctions

```python
def select_year(year):
    
    dataset = df.loc[df.YEAR == year]
    
    return dataset
```

```python
#df = df.append({'A': i}, ignore_index=True)

def append_row(
    ds,
    df_labels,
    
    day_of_week,
    district,
    hour,
    month,
    day,
    night,
    ToNight,
    ToDay,
    temMin,
    temMax,
    temDif,
    precip,
    snow,
    
    g1,
    g2,
    g3,
    g4,
    g5,
    g6,
    g7,
    g8,
    g9
):
    df_labels = df_labels.append(
        {
            'DAY_OF_WEEK':day_of_week, 
            'DISTRICT':district,
            'HOUR':hour, 
            'MONTH':month, 
            'Day':day, 
            'Night':night, 
            'ToNight':ToNight,
            'ToDay':ToDay, 
            'temperatureMin':temMin, 
            'temperatureMax':temMax, 
            'temperatureDifference':temDif,
            'precipitation':precip, 
            'snow':snow, 
            'Motor_Vehicle_Accident_Response':g1, 
            'Larceny':g2,
            'Medical_Assistance':g3, 
            'Simple_Assault':g4, 
            'Violations':g5,
            'Investigate_Person':g6, 
            'Vandalism':g7, 
            'Drug_Violation':g8,
            'Larceny_From_Motor_Vehicle':g9
        },
        ignore_index = True
    ) 
    
    return df_labels   
```

```python
def create_binary_label(group_unique, group):
    
    if (group in group_unique):
        return 1
    else:
        return 0   
```

### 2012

```python
ds = select_year(2012)
```

```python
i = 0
# month
month = 1
while month <= 12:
    
    day_max = ds.DAY.loc[ds.MONTH == month].max()
    day = 1
    
    while day <= day_max:
        
        hour = 0
        
        while hour < 24:
            
            district = 1
            
            while district < 13:
                
                select = (ds.MONTH == month) & (ds.DAY == day) & (ds.HOUR == hour) & (ds.DISTRICT == district)
                
                group_unique = ds.OFFENSE_CODE_GROUP.loc[select].unique()
                                
                if len(ds.loc[select]) != 0:
                    
                    i += 1
                    
                    
                    
                    df_labels = append_row(
                        ds,
                        df_labels,
    
                        day_of_week = float(ds.DAY_OF_WEEK.loc[select].unique()),
                        district = float(ds.DISTRICT.loc[select].unique()),
                        hour = float(ds.HOUR.loc[select].unique()),
                        month = float(ds.MONTH.loc[select].unique()),
                        day = float(ds.Day.loc[select].unique()),
                        night = float(ds.Night.loc[select].unique()),
                        ToNight = float(ds.ToNight.loc[select].unique()),
                        ToDay = float(ds.ToDay.loc[select].unique()),
                        temMin = float(ds.temperatureMin.loc[select].unique()),
                        temMax = float(ds.temperatureMax.loc[select].unique()),
                        temDif = float(ds.temperatureDifference.loc[select].unique()),
                        precip = float(ds.precipitation.loc[select].unique()),
                        snow = float(ds.snow.loc[select].unique()),
    
                        g1 = create_binary_label(group_unique,1),
                        g2 = create_binary_label(group_unique,2),
                        g3 = create_binary_label(group_unique,3),
                        g4 = create_binary_label(group_unique,4),
                        g5 = create_binary_label(group_unique,5),
                        g6 = create_binary_label(group_unique,6),
                        g7 = create_binary_label(group_unique,7),
                        g8 = create_binary_label(group_unique,8),
                        g9 = create_binary_label(group_unique,9)
                    )
                    
                    district += 1
                else:
                    district += 1            
            
                
            hour += 1
        
        day += 1 
    
    month += 1
```
