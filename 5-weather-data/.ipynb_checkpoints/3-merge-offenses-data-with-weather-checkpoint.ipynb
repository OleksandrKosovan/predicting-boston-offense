# Import

```python
import pandas as pd
```

# Read data

```python
w = pd.read_csv('total-weather-for-merge.csv')
```

```python
w.head()
```

```python
w = w.drop('Unnamed: 0', 1)
```

```python
w.snow = pd.to_numeric(w.snow, errors='coerce')
```

```python
w.describe()
```

```python
c = pd.read_csv('crime-boston_1.csv')
```

```python
c.head()
```

```python
c = c.drop('Unnamed: 0',1)
```

```python
w.columns
```

```python
c.OCCURRED_ON_DATE = pd.to_datetime(c.OCCURRED_ON_DATE)
```

```python
c['DAY'] = c.OCCURRED_ON_DATE.dt.day
```

```python
c['temperatureMin'] = 0
c['temperatureMax'] = 0
c['precipitation'] = 0
c['snow'] = 0
```

# Merging

```python
year = 2012
m = 7

while m <=12:
    
    d = 1
    
    while d <= len(w.day.loc[(w.year == year) & (w.month == m)].unique()):
        
        # temperatureMin
        tmin = w.temperatureMin.loc[
            (w.year == year) & (w.month == m) & (w.day == d)
        ]
        
        c.temperatureMin.loc[
            (c.YEAR == year) & (c.MONTH == m) & (c.DAY == d)
        ] = float(tmin)
        
        # temperatureMin
        tmax = w.temperatureMax.loc[
            (w.year == year) & (w.month == m) & (w.day == d)
        ]
        
        c.temperatureMax.loc[
            (c.YEAR == year) & (c.MONTH == m) & (c.DAY == d)
        ] = float(tmax)
        
        
        # precipitation
        p = w.precipitation.loc[
            (w.year == year) & (w.month == m) & (w.day == d)
        ]
        
        c.precipitation.loc[
            (c.YEAR == year) & (c.MONTH == m) & (c.DAY == d)
        ] = float(p)
        
        # snow
        sn = w.snow.loc[
            (w.year == year) & (w.month == m) & (w.day == d)
        ]
        
        c.snow.loc[
            (c.YEAR == year) & (c.MONTH == m) & (c.DAY == d)
        ] = float(sn)
        
        d +=1
    
    m += 1
```
