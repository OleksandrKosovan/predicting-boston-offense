# Import

```python
import numpy as np
import pandas as pd
```

```python
# visualization

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# Data

```python
df_uc = pd.read_csv('data-study/Colleges_and_Universities.csv')
df_np = pd.read_csv('data-study/Non_Public_Schools.csv')
df_p = pd.read_csv('data-study/Public_Schools.csv')
```

```python
df_uc = df_uc[['Latitude','Longitude']]
```

```python
df_uc.describe()
```

```python
df_uc = df_uc[0:56]
```

```python
x = df_uc['Longitude']
y = df_uc['Latitude']
```

```python
# Custom the inside plot: options are: “scatter” | “reg” | “resid” | “kde” | “hex”

sns.jointplot(x, y, kind='scatter')
sns.jointplot(x, y, kind='hex')
sns.jointplot(x, y, kind='kde')
```
